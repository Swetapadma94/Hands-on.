```python
import pandas as pd
```

```python
df=pd.read_csv(r'E:\Handson\Multicollinearity-master\data\Advertising.csv',encoding='latin1')
```

```python
df.head(5)
```

```python
X = df[['TV', 'radio','newspaper']]
y = df['sales']
df.head()
```

```python
X
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python

## fit a OLS model with intercept on TV and Radio

model= sm.OLS(y, X).fit()
```

```python
model
```

```python
model.summary()
```

# As this model is having less standard errors also P value is less than 0.05 and bychecking correlation using corr() function we came to a conclusion that multi collinearity issue is not there.

```python
X.iloc[:,1:].corr()
```
